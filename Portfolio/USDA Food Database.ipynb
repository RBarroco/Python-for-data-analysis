```python
nutrients = []

for i in db:
    fnuts = pd.DataFrame(db[i]['nutrients'], index=['value'],
                         columns=['value', 'units', 'description','group'])
    fnuts['id'] = db[i]['id']
    nutrients.append(fnuts)

nutrients = pd.concat(nutrients, ignore_index=True)
```

```python
nutrients = nutrients.drop_duplicates()
```
